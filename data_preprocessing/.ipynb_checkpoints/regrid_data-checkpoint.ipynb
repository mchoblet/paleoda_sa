### Regridding the SADA product

```python
#upsampling (this is also downsampling! as the calib data sets are of higher resolution)
ccsm=xr.open_dataset('/home/mchoblet/paleoda/data/model_data/CCSM_d18O_851_1850.nc')
ccsm=ccsm.resample(time='YS').mean('time')
new_folder='/home/mchoblet/validation_data/sa_project/upsampled/'
path='/home/mchoblet/validation_data/sa_project/scPDSI_SADA_annual-djf_1400_2000.nc'
var='scpdsi'

current_mod=xr.open_dataset(path)
current_mod['time']=xr.cftime_range(start='1400',end='2000',freq='YS',calendar='noleap')
current_mod['lon']=current_mod['lon']%360

#adjust grid of ccsm

lon_0=current_mod['lon'][0].values.item()
lon_1=current_mod['lon'][-1].values.item()
lat_0=current_mod['lat'][0].values.item()
lat_1=current_mod['lat'][-1].values.item()
#mr=0
#ccsm=ccsm.sel(lon=slice(lon_0-mr,lon_1+mr),lat=slice(lat_0+mr,lat_1-mr))
ccsm=ccsm.sel(lon=slice(lon_0,lon_1),lat=slice(lat_0,lat_1))
regridder=xe.Regridder(current_mod,ccsm,'bilinear',periodic=False) #set boundary conditions to non periodic for this small portion   

file=current_mod
regrided=regridder(file)
#regrided=regrided.to_dataset(name=var)
regrided[var].attrs=file[var].attrs
tail=os.path.split(path)[1]
new_path=new_folder+tail
regrided.to_netcdf(new_path)

regrided['scpdsi'][0].plot()
#regrided[0].plot()
```
